1. Задание
Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)

1)	Найдите их сумму. (на листочке)

(10, 10, 10) + (0, 0, -10) = (10, 10, 0)

2)	Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
def len_vector(vector):
    return np.sqrt(sum([x**2 for x in vector]))
```

```python
tmp = np.array([2, 5, 3])
len_vector(tmp)
```

2. Задание (на листочке)
Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 21)
y = 3*x + 1
y2 = (-1/3)*x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.ylabel('x')
plt.ylabel('y')
```

Прямые не кажутся перпендикулярными, т.к. размерностей осей х и у разные.

3. Задание (в программе
Напишите код на Python, реализующий построение графиков:
1.	окружности,
2.	эллипса,
3.	гиперболы.

```python
# x**2 + y**2 = r**2
r = 5
x = np.linspace(-5, 5, 1000)
y = np.sqrt(-x**2 + r**2)
plt.plot(x, y, 'b')
plt.plot(x, -y, 'b')
plt.gca().set_aspect('equal')
```

```python
# x**2/a**2 + y**2/b**2 = 1
a = 2
b = 0.8
x = np.linspace(-2, 2, 1000)
y = np.sqrt(b**2 - ((x*b)**2)/a**2)
plt.plot(x, y, 'b')
plt.plot(x, -y, 'b')
plt.gca().set_aspect('equal')
```

```python
# x**2/a**2 - y**2/b**2 = 1
a = 4
b = 2
x = np.linspace(-10, 10, 1500)
y = np.sqrt(((x*b)**2)/a**2 - b**2)
plt.plot(x, y, 'b')
plt.plot(x, -y, 'b')
plt.gca().set_aspect('equal')
```
