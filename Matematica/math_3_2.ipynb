1. Задание (в программе)
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
x = np.linspace(-np.pi, np.pi, 201)
k1 = 2
k2 = 1.5
a1 = 1
a2 = 2
b1 = 3
b2 = 1
plt.plot(x, k1*np.cos(x - a1) + b1)
plt.plot(x, k2*np.cos(x - a2) + b2)
plt.ylabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

3.2. Напишите код, который будет рисовать график окружности в полярных координатах.
3.3. Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

```python
import math
def pol2decart(rho, phi):
    x = rho * math.cos(math.radians(phi))
    y = rho * math.sin(math.radians(phi))
    return(x, y)
pol2decart(5,5)
```

```python
fig = plt.figure(figsize=(10., 10.))
phi = np.arange(0, 2*np.pi, 0.01)
rho = 5*np.cos(phi)

ax1 = fig.add_subplot(231, projection='polar')
ax1.plot(rho, phi)
ax1.grid(True)
plt.show()
```
